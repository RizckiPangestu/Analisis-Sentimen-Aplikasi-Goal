```python
#Library Scrapping
from google_play_scraper import app, Sort, reviews
import pandas as pd
import numpy as np
```

```python
app_id = "com.freerange360.mpp.GOAL"
```

```python
result, continuation_token = reviews(
    app_id,
    lang='en',
    country='id',
    sort=Sort.NEWEST,
    count=30000,
    filter_score_with=None
)
```

```python
df_data_goal = pd.DataFrame(np.array(result),columns=['review'])
df_data_goal = df_data_goal.join(pd.DataFrame(df_data_goal.pop('review').tolist()))
df_data_goal.head()
```

```python
df_data_goal.rename(columns={
    'content':'review',
    'score':'rating',
    'at':'review_date'
}, inplace=True)

df_data_goal.head()
```

```python
#Mengambil kolom-kolom penting
df_data_goal_final = df_data_goal[['reviewId','userName', 'review', 'rating', 'review_date']]
df_data_goal_final.head()
```

# **Sentimen analisis pada aplikasi Goal.com**

```python
#Library untuk sentimen analisis
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
from google.colab import files
```

```python
sia = SentimentIntensityAnalyzer()

def calculate_sentiment(review):
    sentiment_score = sia.polarity_scores(review)
    return sentiment_score['compound']

def categorize_sentiment(score, rating):
  if score > 0.05:
    if rating >= 4:
      return 'Positive'
    elif rating == 3:
      return 'Mixed Positive'
    else:
      return 'Mixed Negative'
  elif score < -0.05:
    if rating <= 2:
      return 'Negative'
    elif rating == 3:
      return 'Mixed Negative'
    else:
      return 'Mixed Positive'
  else:
    if rating >= 4:
      return 'Positive'
    elif rating <= 2:
      return 'Negative'
    else:
      return 'Neutral'

def sentiment_bucket(score):
  if score > 0.5:
    return '0.5 to 10'
  elif 0 <= score <= 0.5:
    return '0 to 0.49'
  elif -0.5 <= score < 0:
    return '-0.49 to 0'
  else:
    return '-1 to -0.5'
```

```python
df_data_goal_final_clean['sentiment_score'] = df_data_goal_final_clean['review'].astype(str).apply(calculate_sentiment)
df_data_goal_final_clean.head()
```

```python
df_data_goal_final_clean['sentiment_category'] = df_data_goal_final_clean.apply(
    lambda row: categorize_sentiment(row['sentiment_score'], row['rating']), axis=1)
df_data_goal_final_clean.head()
```

```python
df_data_goal_final_clean['sentiment_bucket'] = df_data_goal_final_clean['sentiment_score'].apply(sentiment_bucket)
df_data_goal_final_clean.head()
```

```python
#Menghitung jumlah sentimen score bernilai 0
jumlah_sentimen_score_nol = (df_data_goal_final_clean['sentiment_score'] == 0).sum()
print("Jumlah sentimen score bernilai 0:", jumlah_sentimen_score_nol)

#Menghitung total data
jumlah_data = len(df_data_goal_final_clean)
print("Total data:", jumlah_data)

#Menghitung persentase
persentase = (jumlah_sentimen_score_nol / jumlah_data) * 100
print("Persentase sentimen score bernilai 0:", persentase, "%")
```
